### 1. Haga un programa en Python que reciba de parametro un numero de segundos y que despliegue en pantalla, las horas, minutos y segundos que representan.

Ejemplo: el numero de segundos recibidos fuera: 2743 entonces su programa deberıa desplegar:
Horas: 0, Minutos: 45 , Segundos: 43

```python
def fn_conversionSegundos(segundos):
    horas =segundos // 3600
    min_sobran = segundos % 3600
    minutos = min_sobran // 60
    segundos_sobran = min_sobran % 60
    return horas, minutos, segundos_sobran


while True:
    try:
        segInput = int(input("ingrese una cantidad de segundos: "))
    except:
        print("no ingreso un numero valido, intente de nuevo...")
        continue
        
    horas, minutos, segundos = fn_conversionSegundos(segInput)
    print(f"Horas: {horas} Minutos: {minutos} Segundos: {segundos}")
    break
```

despliegue cuatro patrones triangulares.

### 3. Haga un programa en Python que la fecha de nacimiento de una persona 
Puede ingresar por separado cada valor, es decir pedir primero dıa, luego mes y luego año, despliegue la edad de la persona en años, meses y dias. Su resultado puede ser aproximado.

```python
from datetime import datetime

def fn_calcularEdad(dia,mes, anio):
    fecha = datetime.today()
    edad = fecha.year - anio
    
    if fecha.month<mes:
        edad =edad-1
    elif fecha.month==mes and fecha.day<dia:
        edad=edad-1
    
    return edad

while True:
    try:
        dia= int(input("ingrese su dia de nacimiento: "))
        mes= int(input("ingrese su mes de nacimiento: "))
        anio= int(input("ingrese su año de nacimiento: "))

    except:
        print("no ingreso un numero entero ")
        continue

    if(dia>31 or dia<=0):
        print("el dia ingresado no es valido, intente de nuevo")
        continue

    if(mes>12 or mes<=0):
        print("el dia ingresado no es valido, intente de nuevo")
        continue

    if(anio<=0):
        print("el dia ingresado no es valido, intente de nuevo")
        continue
        
    print(f"Su edad es: {fn_calcularEdad(dia,mes, anio)}")
    break

```

### 4. Escriba una funcion en Python que dada una tupla de elementos (numeros o letras) esta se convierta a un string.

Tomar nota del siguiente ejemplo:
1. Si la entrada fuera: (10, 20, 40, 5, 70)
1. La salida deberia ser: 102040570

```python
def fn_tupla_string(tupla):
    resultado=""
    for valor in tupla:
        resultado += str(valor)
    return resultado

tupla = (10, 20, 40, 5, 70)
print(f"Resultado: {fn_tupla_string(tupla)}")
    
```

### 5. Cree un programa en Python que dada una lista de tuplas, elimine las tuplas que estan vacias.
Tomar nota del siguiente ejemplo:
- Si la lista fuera: [(), (), (’X’,), (’a’, ’b’), (’a’, ’b’, ’c’), (’d’)]
- El resultado deber´ıa ser: [(’X’), (’a’, ’b’), (’a’, ’b’, ’c’), ’d’]

```python
def fn_limpia_lista(lista):
    resultado =[]
    for valor in lista:
        if(len(valor)>0):
            resultado.append(valor)
            
    return resultado
    
lista =[(), (), ("X",), ("a", "b"), ("a", "b", "c"), ("d")]

print(f"Resultado:  {fn_limpia_lista(lista)}")
```

### 6. Cree una funcion en Python que dada una tupla de tuplas con numeros, produzca una tupla con el promedio de cada tupla individual.
debe usar comprehensions para desarrollar este ejercicio. 
Tomar nota del siguiente ejemplo:
- Si la tupla de entrada fuera: ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 32))
- El resultado deber´ıa ser: (10.5, 44, 58)

```python
def fn_promedioTuplas(tupla):
    resultado= tuple(sum(tp)/len(tp) for tp in tupla)
    return resultado
tupla =  ((10, 10, 10, 12), (30, 45, 56, 45), (81, 80, 39, 3))


print(f"El resultado es: {fn_promedioTuplas(tupla)}")
```
